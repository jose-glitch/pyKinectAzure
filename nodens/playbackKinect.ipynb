# Playback RADAR + AZURE KINECT

## Necessary modules

```python
import sys
import cv2
import math

sys.path.insert(1, '../')
import pykinect_azure as pykinect
import matplotlib.pyplot as plt
```

# Load data from Kinect

### Read configuration

```python
video_filename = "QMofficeSeries1.mkv"

# Initialize the library, if the library is not found, add the library path as argument
pykinect.initialize_libraries()
# Start playback
playback = pykinect.start_playback(video_filename)
playback_config = playback.get_record_configuration()
playback_recLength = playback.get_recording_length()/1e6
playback_frames = playback_recLength*30
# First Image:
ret, capture = playback.update()
ret_color, color_image = capture.get_color_image() 
# print(playback_config)
```

### Read data

```python
from IPython.display import display, clear_output

fig, ax1 = plt.subplots()
myimage = ax1.imshow(color_image)

for n in range(math.floor(playback_frames)-1):
    ax1.cla()
    # Get camera capture
    ret, capture = playback.update()
    # Get color image
    ret_color, color_image = capture.get_color_image()
    # ret_color, color_image = capture.get_transformed_color_image()
    # Get the colored depth
    # ret_depth, depth_color_image = capture.get_colored_depth_image()
    # Plot the image
    # combined_image = cv2.addWeighted(color_image[:, :, :3], 0.7, depth_color_image, 0.3, 0)
    ax1.imshow(color_image)
    display(fig)
    # clear_output(wait=True)
    # plt.pause(0.0001)
```
